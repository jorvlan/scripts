# Loading and inspecting the pupil labs pilot  data

This notebook is to see whether I can load and inspect the pupil data as obtained with pupil labs during one of the first pilot sessions of the seated VR task. 

## 1 - Loading the exported data

```python
import pandas as pd
```

```python
pupil_df = pd.read_csv('/Users/jordyvanlangen/data/pupil_positions.csv')
pupil_df
```


# Individual Datums
As we can see the data frame contains many columns, representing all the columns present in the CSV file. In the sample recording we used the 3D gaze mapping mode and most of the values are concerned with the state of the 3D eye model used in this mode. We will now take a closer look at the subset of values most important for the next step, where we want to visualize pupil diameter and pupil location:

- timestamp - The timestamp at which the sample was recorded.
- id - Identifies the eye to which the sample belongs to. By default samples with id = 0 belong to the right eye, while samples with id = 1 belong to the left eye (from the wearer's POV!).
- confidence - Confidence in the pupil detection.
- norm_pos - The normalized position of the pupil in the eye image. (0,0) is bottom left corner of the eye image frame; (1,1) is the top right corner of the eye image frame.
- diameter - The pupil diameter in pixels.

For more information on the other available values please refer to the data format documentation.

Let's take a look at a few values from our sample data.

```python
eye_r_df = pupil_df[pupil_df.id == 0]
eye_l_df = pupil_df[pupil_df.id == 1]
pd.options.display.float_format = '{:.3f}'.format

print("Right eye data:")
print(eye_r_df[['timestamp', 'id', 'confidence', 'norm_pos_x', 'norm_pos_y', 'diameter']][:10])

print("Left eye data:")
print(eye_l_df[['timestamp', 'id', 'confidence', 'norm_pos_x', 'norm_pos_y', 'diameter']][:10])
```

# 2 - Visualize the data

Now that we have the data available, we can visualize it

```python
import matplotlib.pyplot as plt
```

Plot the pupil diameter as a function of time:

```python
plt.figure(figsize = (16, 5))
plt.plot(eye_r_df['timestamp'], eye_r_df['diameter'])
plt.plot(eye_l_df['timestamp'], eye_l_df['diameter'])
plt.legend(['eye_r', 'eye_l'])
plt.xlabel('Timestamps [s]')
plt.ylabel('Diameter [pixels]')
plt.title('Pupil Diameter')
```

As we can see there are unreasonable spikes in the data. These spikes are due to false positive pupil detections typically occuring during blinks. We can filter out most of the erroneous values using the confidence measure.

```python
eye_r_high_conf_df = eye_r_df[eye_r_df['confidence'] > .95]
eye_l_high_conf_df = eye_l_df[eye_l_df['confidence'] > .95]

plt.figure(figsize = (16, 5))
plt.plot(eye_r_high_conf_df['timestamp'], eye_r_high_conf_df['diameter'])
plt.plot(eye_l_high_conf_df['timestamp'], eye_l_high_conf_df['diameter'])
plt.legend(['eye_r', 'eye_l'])
plt.xlabel('Timestamps [s]')
plt.ylabel('Diameter [pixels]')
plt.title('Pupil Diameter (conf > .95 %)')
```
